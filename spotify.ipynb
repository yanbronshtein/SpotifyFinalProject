```python
model=Sequential()
model.add(Dense(11870, input_dim=23743, activation=LeakyReLU(alpha=0.01)))
model.add(Dense(5935, activation=LeakyReLU(alpha=0.01)))
model.add(Dense(2967, activation=LeakyReLU(alpha=0.01)))
model.add(Dense(1186, activation=LeakyReLU(alpha=0.01)))
model.add(Dense(11870, activation=LeakyReLU(alpha=0.01)))
model.add(Dense(23743, activation='softmax'))
model.compile(loss=tf.keras.losses.categorical_crossentropy,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])
print(model.summary())
```
